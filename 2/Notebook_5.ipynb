```python
import seaborn as sns
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# missing Values
print(titanic.isnull().sum())
```

```python
data = titanic[['age', 'fare', 'pclass', 'sex']]
data = pd.get_dummies(data, drop_first=True) # convierte la variable 'sexo' en una columna binaria
data.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
print("Datos originales")
print(data.head(30))
print("Datos imputados")
print(data_imputed.head(30))
```

```python
import kagglehub
import os
import shutil
import pandas as pd

path = kagglehub.dataset_download("kumarajarshi/life-expectancy-who")

print("Path to dataset files:", path)

drive_path = "/content/drive/MyDrive/AI/Files/"

os.makedirs(drive_path, exist_ok=True)

for item in os.listdir(path):
    s = os.path.join(path, item)
    d = os.path.join(drive_path, item)
    if os.path.isdir(s):
        shutil.copytree(s, d, dirs_exist_ok=True)
    else:
        shutil.copy2(s, d)

print(f"Dataset copied to: {drive_path}")
```

```python
life_expectancy_path = "/content/drive/MyDrive/AI/Files/Life Expectancy Data.csv"

import pandas as pd
life_expectancy_df = pd.read_csv(life_expectancy_path)

print("Primeras filas del dataset de Life Expectancy:")
display(life_expectancy_df.head(10))

print("\nResumen de valores faltantes antes de la imputaci√≥n:")
print(life_expectancy_df.isnull().sum())
```

```python
life_expectancy_numeric = life_expectancy_df.select_dtypes(include=['number'])
imputer_alt1 = IterativeImputer(max_iter=10, random_state=42)

life_expectancy_imputed_alt1 = pd.DataFrame(imputer_alt1.fit_transform(life_expectancy_numeric), columns=life_expectancy_numeric.columns)

print("Resumen de valores faltantes antes de la imputaci√≥n (columnas num√©ricas):")
print(life_expectancy_numeric.isnull().sum())

print("\nResumen de valores faltantes despu√©s de la imputaci√≥n (columnas num√©ricas):")
print(life_expectancy_imputed_alt1.isnull().sum())

print("\nPrimeras filas del dataset de Life Expectancy despu√©s de la imputaci√≥n (versi√≥n alternativa 1):")
display(life_expectancy_imputed_alt1.head(10))
```

## Ejercicio 2:

Use MICE on the Planets (seaborn) dataset to fill in missing values.

```python
import seaborn as sns
import pandas as pd
import os
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

planets = sns.load_dataset('planets')
planets.head()
```

```python
print(planets.isnull().sum())
```

```python
planets_cols_alt2 = planets[['orbital_period', 'mass', 'distance']]
planets_cols_alt2.head()
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
imputer_alt2 = IterativeImputer(max_iter=20, random_state=1)

planets_imputed_alt2 = pd.DataFrame(imputer_alt2.fit_transform(planets_cols_alt2), columns=planets_cols_alt2.columns)

print("Comparaci√≥n de los primeros 5 valores originales y imputados:")
for col in planets_cols_alt2.columns:
    if planets_cols_alt2[col].isnull().any():
        print(f"\nColumna: {col}")
        print("Original:")
        print(planets_cols_alt2[col].head().to_string())
        print("Imputado:")
        print(planets_imputed_alt2[col].head().to_string())

print("\nPrimeras filas del dataset de planets despu√©s de la imputaci√≥n (versi√≥n alternativa 2):")
display(planets_imputed_alt2.head())
```
